```python
import requests
import json
```

```python
api_url = f'https://opendata-download-metfcst.smhi.se/api/category/pmp3g/version/2/geotype/point/lon/16.158/lat/58.5812/data.json'
```

```python
smhi = requests.get(api_url)
```

```python
data = smhi.json()
```

```python
print(data)
```

```python
pretty_data = json.dumps(data, indent= 4)
```

```python
print(pretty_data)
```

```python
#tempreture = data['timeSeries']["validTime"]["parameters"]["name": "t"]

# Assuming data['timeSeries'] is a list of dictionaries
#temperature = [item for item in data['timeSeries'] if item.get('name') == 't']

# Iterate through the timeSeries list
for item in data['timeSeries']:
    valid_time = item['validTime']
    
    # 'parameters' is assumed to be a list, we iterate over it to find the one where name is 't'
    for param in item['parameters']:
        if param.get('name') == 't':
            temperature = param['values']  # Get the temperature values
            print(f"Temperature at {valid_time}: {temperature}")

```

```python
# Sort the timeSeries list by 'validTime' to get the latest time (assuming validTime is in ISO 8601 format)
latest_entry = max(data['timeSeries'], key=lambda x: x['validTime'])

# Find the temperature parameter where 'name' is 't'
temperature = next((param['values'] for param in latest_entry['parameters'] if param.get('name') == 't'), None)

# Output the latest validTime and corresponding temperature
if temperature:
    print(f"Latest valid time: {latest_entry['validTime']}")
    print(f"Latest temperature: {temperature[0]}")
else:
    print("Temperature data ('t') not found.")
```
